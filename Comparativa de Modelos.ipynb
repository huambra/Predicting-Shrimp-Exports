##### Comentarios finales sobre la comparación de modelos (SARIMA, RF y LSTM)

En este cuadro final se han incluido, además de los distintos SARIMA (con validación cruzada, transformaciones, recorte de datos, etc.), modelos de Random Forest y redes LSTM. Los puntos clave son:
##### Escalas distintas en LSTM (MAE, RMSE):

Al normalizar la serie para entrenar la LSTM, los valores de MAE y RMSE quedan en una escala muy reducida (por ejemplo, 0.148). Esto dificulta la comparación directa con los modelos SARIMA o RF, que operan en la escala original.
Sin embargo, el MAPE sirve de referencia común, ya que es un error porcentual.
##### SARIMA vs. RF

En general, los modelos RF Base y RF Fine-Tuned muestran MAPE bastante altos para China (24%–25%) y valores también elevados para UE (~17%–18%).
Esto indica que, en esta configuración, Random Forest no está capturando bien la dinámica de la serie de tiempo, resultando en predicciones más alejadas que la mayoría de los SARIMA.
##### SARIMA vs. LSTM

Para China, la LSTM Base ~17.45% y la LSTM Fine-Tuned ~17.25% de MAPE compiten con algunos SARIMA de error similar (p. ej., SARIMA 2020+ con 17.01%).
Para USA, la LSTM Fine-Tuned baja el MAPE a ~18.66% (mejorando sobre la LSTM Base ~21.97%), aunque sigue más alto que varios SARIMA “base” (por ejemplo, ~13.15% con SARIMA Base).
En el caso de la UE, la LSTM Base marca ~20.03% de MAPE y la Fine-Tuned ~23.24%, lo que implica un empeoramiento tras el ajuste (posible sobreajuste o mala generalización).
##### Rendimiento diferencial por país

China: SARIMA Base (12.38% MAPE) continúa relativamente fuerte. LSTM y algunos SARIMA transformados andan entre ~14–17%. Random Forest, con ~24–25%, queda rezagado.
USA: El SARIMA Base (13.15%) sobresale en comparación con la mayoría de modelos alternativos. La LSTM Fine-Tuned mejora frente a su versión base (de ~21.97% a ~18.66%), pero no alcanza el nivel del SARIMA Base.
UE: Los SARIMA con transformaciones (Log, Box-Cox) y el Baseline (~9.8%–11.3% MAPE) se comportan bien; LSTM mantiene errores porcentuales en torno al 20% o más, y RF se queda en torno a 17–18%.
##### Conclusiones generales

Los modelos SARIMA siguen posicionados como competidores sólidos: SARIMA Base o con ligeros ajustes suelen ofrecer los menores MAPE en la mayoría de países.
Random Forest no presenta mejoras significativas en estos datos, quizás necesitando otras configuraciones o más features exógenas para capturar la dinámica temporal.
LSTM muestra potencial de mejora, sobre todo en USA, donde la versión Fine-Tuned reduce el MAPE. Sin embargo, en otros casos (UE) no supera a SARIMA, probablemente requiriendo mayor experimentación en hiperparámetros, arquitectura o reescalado.
En suma, no hay un “ganador absoluto”: cada país mantiene patrones distintos y, para China y USA, muchos de los modelos avanzados quedan por detrás del SARIMA Base en términos de error porcentual. Las LSTM tienen margen de maniobra, pero se requiere un ajuste delicado para adaptarse a la elevada variabilidad de las series. Por su parte, Random Forest, en su configuración actual, no alcanza el desempeño de SARIMA o LSTM.
